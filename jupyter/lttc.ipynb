```python

from datetime import date, timedelta
import datetime

def daterange(start_date, end_date):
    for n in range(int((end_date - start_date).days)):
        yield start_date + timedelta(n)



# Create a rolling 30-day grap

start_date = date(2022, 9, 1)
end_date = date(2023, 9, 5)

dfRolling = pd.DataFrame(columns=['Date', 'Avg', 'Count'])



# Yes this is a very stupid way to do ths..
for single_date in daterange(start_date, end_date):
    count = 0
    lttc  = 0
    for _, row in df.iterrows():
        dtRow = row['Merged'].to_pydatetime().date()
        datediff = single_date - dtRow

        if datediff.days >= 0 and datediff.days < 30:             
            count = count + 1
            lttc  = lttc + row['LTTC']
#            print("" + str(dtRow) + " -> " + str(single_date) + " : " + str(datediff) + " YES")
#        else:
#            print("" + str(dtRow) + " -> " + str(single_date) + " : " + str(datediff) + " NO")
    if count != 0 :    
        dfRolling = pd.concat([dfRolling, pd.DataFrame([{'Date': single_date, 'Avg': lttc/count, 'Count': count}])], ignore_index=True )

dfRolling
#    print(single_date.strftime("%Y-%m-%d")



```
